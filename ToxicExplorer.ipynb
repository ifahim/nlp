```python
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r"[\w']+")
tokenizer.tokenize('Eighty-seven miles to go, yet.  Onward!')
```

```python
tokenizer.tokenize("I'm too tired. I don't know which is best")
```

```python
import re
```

```python
def get_word_set(comments):
    word_set = set()
    for i, comment in comments.iteritems():
        words = tokenizer.tokenize(comment)
        for word in words:
            word = word.lower()
            word = re.sub('newline_token', '', word)
            word_set.add(word.lower())
    return word_set
```

```python
toxic_set = get_word_set(toxic_comments)
non_toxic_set = get_word_set(non_toxic_comments)
```

```python
toxic_words = set()
for word in toxic_set:
    if word not in non_toxic_set:
        toxic_words.add(word)
```

```python
len(toxic_words)
```

```python
sorted(toxic_words)
```

```python
def toxic_word_count(comment, tokenizer, toxic_words):
    words = tokenizer.tokenize(comment)
    toxic_count = 0
    for word in words:
        if word in toxic_words:
            toxic_count += 1
    return toxic_count
            
```

```python
comment = "blah blah sucks stupid jerk"
toxic_word_count(comment, tokenizer, toxic_words)
```

```python
pd.options.mode.chained_assignment = None
```

```python
scored['toxic_words'] = scored['comment'].apply(lambda x: toxic_word_count(x, tokenizer, toxic_words))
```
